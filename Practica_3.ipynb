EJERCICIO 1: Resume lo mas que puedas en una celda los pasos imprescindibles para aplicar PCA "a mano". Comenta cada linea.

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

iris = datasets.load_iris()
X = iris.data

# 1er Paso: Calcular la media
m = np.mean(X, axis=0)

# 2do Paso: Centrar los datos restando la media
Xm = X - m[np.newaxis, :]

# 3er Paso: Calcular la matriz de covarianza
cov_matrix = np.dot(Xm.T, Xm) / Xm.shape[0]

# 4to Paso: Calcular los autovalores y autovectores de la matriz de covarianza
evals, evecs = np.linalg.eigh(cov_matrix)

# 5to Paso: Proyectar los datos centrados en el nuevo espacio PCA
Xpca = np.dot(evecs.T, Xm.T).T

print(Xpca)
```

EJERCICIO 3: Aplica PCA usando sklearn a los datos de Boston. Explica cada paso.

```python
from sklearn.decomposition import NMF
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
X = iris.data

# Creando el modelo de Factorización de Matrices No Negativas
nmf = NMF(n_components=2, init='random', random_state=42)

# Ajustando y transformando los datos
X_nmf = nmf.fit_transform(X)

# Imprimiendo los resultados
print(X_nmf)
```
